```python
import urllib.request
url="https://raw.githubusercontent.com/sarahbismuth/DOCP3/master/tournamentinfo.txt"
data=urllib.request.urlopen(url).read()
```

```python
data
```

```python
data = str(data).replace('\\r', '').replace('\\n', '\n')
```

```python
import re
pattern=r"([0-9]+)\s\|\s([A-Z-\s]+)\s*\|([0-9.]+)\s*\|[A-Z\s]*([0-9]*)?\|[A-Z\s]*([0-9]*)?\|[A-Z\s]*([0-9]*)?\|[A-Z\s]*([0-9]*)?\|[A-Z\s]*([0-9]*)?\|[A-Z\s]*([0-9]*)?\|[A-Z\s]*([0-9]*)?\|\s*([A-Z]{2})\s*\|\s[0-9]*\s....\s*([0-9]+)"
```

```python
re.match(pattern, data)
```

```python
players=[]
for item in re.finditer(pattern,str(data),re.MULTILINE):
    print(list(item.groups()))
    players.append(item.groups())
```

```python
import pandas as pd
import numpy as np
```

```python
colnames=["player_id","player_name","total_points","round 1","round 2","round 3","round 4","round 5","round 6","round 7","state","pre_rating"]

df=pd.DataFrame(players,columns=colnames)
df=df.set_index("player_id")
```

```python
df.head()
```

```python
df.iloc[1]
```

```python
def get_score(x):
    if x == '':
        return None
    return int(df.iloc[int(x) - 1]["pre_rating"])
```

```python
df["round 2"]=df["round 2"].apply(get_score)
```

```python
df["round 3"]=df["round 3"].apply(get_score)
df["round 4"]=df["round 4"].apply(get_score)
df["round 5"]=df["round 5"].apply(get_score)
df["round 6"]=df["round 6"].apply(get_score)
df["round 7"]=df["round 7"].apply(get_score)
```

```python
df["op_avg"] = df[["round 1", "round 2", "round 3", "round 4", "round 5", "round 6", "round 7"]].mean(axis=1)
```
